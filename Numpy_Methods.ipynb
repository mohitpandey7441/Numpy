# Round
The numpy.round() function rounds the elements of an array to the nearest integer or to the
specified number of decimals

```python
# Round to the nearest integer
arr = np.array([1.2, 2.7, 3.5, 4.9])
rounded_arr = np.round(arr)
print(rounded_arr)
```

```python
# Round to two decimals
arr = np.array([1.234, 2.567, 3.891])
rounded_arr = np.round(arr, decimals=2)
print(rounded_arr)
```

```python
#randomly
np.round(np.random.random((2,3))*100)
```

# Floor
The numpy.floor() function returns the largest integer less than or equal to each element of an
array.


```python
# Floor operation
arr = np.array([1.2, 2.7, 3.5, 4.9])
floored_arr = np.floor(arr)
print(floored_arr)
```

```python
np.floor(np.random.random((2,3))*100) # gives the smallest integer
```

# 7. Boolean & Logical Functions

```python
# Check if all element are true
# all
logical_test = np.array([True, True, True]) 
all_true = np.all(logical_test)  # Check if all are True
print("All elements True:", all_true)
```

#  Set Operations

```python
# Intersection of two arrays
set_a = np.array([1, 2, 3, 4])
set_b = np.array([3, 4, 5, 6])
intersection = np.intersect1d(set_a , set_b)
print("Intersection of a and b is:",intersection)
```

# Speed of List Vs Numpy

# List

```python
# Element-wise addition
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]
c = []
import time 
start = time.time()
for i in range(len(a)):
    c.append( a[i] + b[i])

print(time.time()-start)    
```

# Numpy

```python
import numpy as np
a = np.arange(10000000)
b = np.arange(10000000,20000000)
start = time.time()
c = a + b
print(time.time()-start)
```
